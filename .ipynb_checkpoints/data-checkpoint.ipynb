```python
# Read and Parse XML file downloaded from UNICEF Activity File - Albania

import xml.etree.ElementTree as ET
import pandas as pd

tree = ET.parse('data/iatiActivity_Albania.xml')
root = tree.getroot()
```

```python
# List all element names in xml file:

element_names = [element.tag for element in root.iter()]

# Print the list of element names
print("List of Element Names:")
for name in element_names:
    print(name)
```

```python
# Check if the iati-activity element includes the description element
has_description = bool(root.find('.//iati-activity/description'))

has_description
```

```python
# Check the structure of the description element
description_element = root.find('.//iati-activity/description')

# Check if the type attribute exists
type_attribute = description_element.get('type', None)

type_attribute
```

```python
# Function to check if description elements contain narrative content
def check_description_content(xml_element):
    for description_element in xml_element.findall('.//iati-activity/description') + xml_element.findall('.//budget-item/description'):
        # Find the narrative sub-element
        narrative_element = description_element.find('narrative')

        # Check if the narrative sub-element exists and has text content
        if narrative_element is None or not narrative_element.text:
            return False  # Narrative content is missing

    return True  # All description elements contain narrative content

# Check if all description elements contain narrative content
all_description_content = check_description_content(root)
all_description_content


```
